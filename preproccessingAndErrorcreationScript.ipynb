```python
class preprocessing:

  def normalize(sentence):
    '''
    Argument:
        string of words
    return:
        string of words but standardize the words
    '''
    sentence = re.sub("[إأآا]", "ا", sentence)
    sentence = re.sub("ى", "ي", sentence)
    sentence = re.sub("ؤ", "ء", sentence)
    sentence = re.sub("ئ", "ء", sentence)
    sentence = re.sub("ة", "ه", sentence)
    sentence = re.sub("گ", "ك", sentence)
    sentence = re.sub(r'[^أ-ي\s]', "", sentence)

    return sentence

  def remove_punkt(sentence):
    '''
      Argument:
          string of words
      return:
          string of words but after removing punctuations
    '''
    sentence = re.sub(r'[^\w\s]',"",sentence)
    return sentence



  def remove_spaces_and_single_chars(sentence):

    sentence = re.sub(r'\b\w\b', '', sentence)
    sentence = re.sub(r'\s{2,}', ' ', sentence)

    return sentence

  def preprocess(sentence):
    '''
    Argument:
        string of words
    return:
        string of words after cleaning
    '''
    clean = preprocessing.normalize(sentence)
    clean = preprocessing.remove_punkt(clean)
    clean = preprocessing.remove_spaces_and_single_chars(clean)
    return clean
```

```python
df['text'] = df['text'].apply(preprocessing.preprocess)
```

```python
import random
import numpy as np

def random_insertion(word):
    if len(word) < 2:
        return word
    random_idx = random.randint(1, len(word)-1)
    return word[:random_idx] + random.choice('ئ') + word[random_idx:]

def random_deletion(word):
    if len(word) < 2:
        return word
    random_idx = random.randint(1, len(word)-1)
    return word[:random_idx] + word[random_idx+1:]

def random_swap(word):
    if len(word) < 2:
        return word
    idx = list(range(len(word)))
    i1, i2 = random.sample(idx, 2)
    word_lst = list(word)
    word_lst[i1], word_lst[i2] = word_lst[i2], word_lst[i1]
    return ''.join(word_lst)

def introduce_errors(text, p):
    words = text.split()
    num_errors = np.random.binomial(len(words), p)
    if num_errors == 0:
        return text

    modified_words = set()
    while len(modified_words) < num_errors:
        index = random.randint(0, len(words) - 1)
        if index not in modified_words:
            error_type = np.random.choice(['insertion', 'deletion', 'swap'], p=[1/3, 1/3, 1/3])
            if error_type == 'insertion':
                words[index] = random_insertion(words[index])
            elif error_type == 'deletion':
                words[index] = random_deletion(words[index])
            elif error_type == 'swap':
                words[index] = random_swap(words[index])
            modified_words.add(index)

    return ' '.join(words)

df['sentence_with_errors'] = df['text'].apply(lambda x: introduce_errors(x, 0.25))  # introduce errors with probability 0.1
```
